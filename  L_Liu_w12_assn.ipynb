```python
import xml.etree.ElementTree as et 
```

```python
xtree = et.parse("books.xml")
xroot = xtree.getroot() 

df_cols = ["title", "pages", "publisher", "author"]
rows = []

for node in xroot: 
    s_title = node.attrib.get("title")
    s_pages = node.attrib.get("pages")
    s_publisher = node.attrib.get("publisher")
    s_author = node.attrib.get("author")
    
    rows.append({"tile": s_title, "pages": s_pages, 
                 "publisher": s_publisher, "author": s_author})

out_df = pd.DataFrame(rows, columns = df_cols)
out_df
```

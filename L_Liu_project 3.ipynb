```python
import re  
import urllib
regex = re.compile('\s+')
import csv
import numpy as np
import pandas as pd
```

```python
#open the text file
with open('tournamentinfo.txt') as file: 
    data = file.readlines()
allRows = [row for row in data] 
allRows= allRows[3:]
allRows
```

```python
# extract the value of player names
pn = re.compile('\|[0-9] | [a-zA-Z]+\s[a-zA-Z]+\s[a-zA-Z]*')
```

```python
# PN = Player Name
PN = pn.findall(str(allRows))
PN
```

```python
# Extract the value of Total Number of Points
tp = re.compile('\|[0-9].[0-9]')
# TP = Total Points
TP = tp.findall(str(allRows))
TP
```

```python
# extract the value of player's state
ps = re.compile('\s[A-Z]{2}\s\|')
```

```python
# PS = Player state
PS = ps.findall(str(allRows))
PS
ps = re.compile('[A-Z]{2}')
PS = ps.findall(str(PS))
PS
```

```python
# Creat DataFrame
Whole = {'Player': PN,'State': PS,'Total Points': TP}
Table = pd.DataFrame(Whole)
Table
```
